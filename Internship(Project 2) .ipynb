```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
```

```python
URL="https://www.bgr.in/gadgets/mobile-phones/search/"
res=[]  #saves the final result
no_of_pages=15
```

```python
for _ in range(no_of_pages):
    r=requests.get(URL) ## used to call the URL and get the contents of the HTML page
    ##Use BeautifulSoup to parse the HTML page
    soup=bts(r.content, 'html5lib') ## if this line causes error install html5lib
    a=soup.findAll("li", class_='mobile-listing') ## get all the specs of mobile specs: figure 1
    ## traverse over the mobile specs links to scrap the specs
    for k in a:
        data={} ## save the specs of mobile in it
        ## open the mobile spec link
        url_tmp=k.find('a')['href']
        mob=requests.get(url_tmp)
        mob=bts(mob.content, 'html5lib') ## parse the html page of mobile spec: fig 2
        ## print(mob)
        info=mob.find_all('span', class_='col-xs-12 col-sm-5 spec-lbl') ## get the col1 from spec table
        desc=mob.find_all('span', class_='col-xs-12 col-sm-7 spec-val') ## get the col2 from spec table
        name=mob.find('h1',{'itemprop':'name'})
        data['name']=name.get_text() ## get the name of the mobile phone
        price=mob.find("span", class_='rsm').get_text() ## get the price of the mobile
        
        for k,v in zip(info, desc):
            data[k.get_text()]=v.get_text()
            res.append(data) ## Store the result into list
    URL=soup.find('a', class_='next pagination')['href'] ## Get the URL to moving to page 2
        
```

```python
pd.DataFrame(res)
```

```python
pd.DataFrame(res).to_csv("Mobile_Data.csv")
```
